# Physics-based Digital Twin: Active Particle Motility

The idea of this notebook is to create a very simple setup and do a very simple simulation of active particles with Neumann boundary condition without collision. An extra ddition to the above setup is now the introduction of low Re fluid flows.


1. Import libraries required for the setup

```python
import numpy as np
import matplotlib.pyplot as plt

print(f"Using numpy version: {np.__version__}")
print(f"Using matplotlib version: {plt.matplotlib.__version__}")
```

2.  Define simulation parameters

```python
# Diffusion coefficient
D = 0.1 # m^2/s
# Time step
dt = 0.01 # s
# No.of steps
n_steps = 5000
# No.of particles
N = 5
# Tube length
L = 5.0  # tube length in m

# Optional: reproducible results
np.random.seed(42)

# Walls control (activate/deactivate)
walls = {
    "left": True,
    "right": True
}

# Wall positions (fixed fractions of L)
wall_positions = {
    "left": 0.0,
    "right": L
}
```

3. For N particle initialize particle position

```python
# Create time array
time = np.arange(n_steps) * dt
# Initialize position array
position = np.zeros((N, n_steps))
```

4. Generate random noise

```python
noise = np.random.normal(0, 1, (N, n_steps))
```

5. Calculation of average fluid velocity along the Tube using Re and other fluid properties:

$$
U = \frac{\mathrm{Re} \cdot \mu}{\rho \cdot D_\mathrm{tube}}
$$

```python
# Fluid properties
rho = 1000     # kg/m^3, density of water
mu = 0.001     # Pa.s (N.s/m^2), viscosity of water
D_tube = 0.01  # m, tube diameter
Re = 0.1       # very low Re

# Compute average velocity
U = Re * mu / (rho * D_tube)
print(f"Average fluid velocity: {U:.5f} m/s")
```

6. Time integration loop

    Using Langevin discretization (Discrete Euler-Maruyama form)

```python
v = U  # use velocity calculated from Re

for i in range(1, n_steps):
    position[:, i] = position[:, i-1] + np.sqrt(2 * D * dt) * noise[:, i] + v * dt

    # Walls reflection
    if walls["left"]:
        position[position[:, i] < 0, i] = -position[position[:, i] < 0, i]
    if walls["right"]:
        position[position[:, i] > L, i] = 2*L - position[position[:, i] > L, i]
```

6. Plot the trajectory

```python
plt.figure(figsize=(8, 4))
for i in range(N):
    plt.plot(time, position[i, :], alpha=0.7)
plt.xlabel("Time")
plt.ylabel("Position")
plt.title("1D Brownian Motion")
plt.grid(True)
plt.show()
```
